```python
import pandas as pd
import sklearn.datasets as ds
import matplotlib.pyplot as plt
```

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="s+", skiprows=22, header=None)
df = (
    pd.concat([raw_df[::2].reset_index(drop=True),
               raw_df[1::2].reset_index(drop=True)], axis=1)
      .dropna(axis=1, how='all')  # elimina columnas vacías
)

X = df.iloc[:, :-1]
y = df.iloc[:, -1]

df.columns = ["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM",
              "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B",
              "LSTAT", "MEDV"]
print(df.head)
```

```python
df.plot(x='CRIM', y='MEDV', kind='scatter')
plt.title('Índice de criminalidad en un eje normal')
plt.show()
```

```python
df.plot(x='CRIM', y='MEDV', kind='scatter', logx=True)
plt.title('Índice de criminalidad en un eje normal')
plt.show()
```
