```python
from transformers import BertTokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
from transformers import pipeline
```

```python
clasificador = pipeline('sentiment-analysis')
```

```python
textos = [ 
    "Me encantó este producto, superó todas mis exprectativas.", 
    "El servicio al cliente fue pésimo y el envío tardó demasiado.", 
    "La calidad es decente, pero no es nada del otro mundo.", 
    "Era justo lo que buscaba.", 
    "Todos deberíamos tener uno." ]
```

```python
texts = [
    "I loved this product, it exceeded all my expectations.",
    "The customer service was terrible and the delivery took too long.",
    "The quality is decent, but nothing out of this world.",
    "It was exactly what I was looking for.",
    "Everyone should have one."
]
```

```python
# Realizar la predicción
prediccion = clasificador(texts)
```

```python
for texto, resultado in zip(texts, prediccion):
    sentimiento = resultado['label']
    puntuacion = resultado['score']
    print('-'*30)
    print(f"Texto: **{texto}**")
    print(f"Sentimiento: **{sentimiento}**")
    print(f"Confianza (score): **{puntuacion:.4f}**")
```
