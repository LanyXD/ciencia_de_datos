```python
# pip install pygwalker
# pip install streamlit

import pandas as pd
import pygwalker as pyg
import seaborn as sb
```

```python
df = pd.read_csv('csv')
df.head()
```

```python
df.dtypes
```

```python
df['date_added'] = df['date_added'].str.strip()
df.head()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
df['date_added_year'] = df['date_added'].dt.year.fillna(0).astype(int)
df['date_added_month'] = df['date_added'].dt.month.fillna(0).astype(int)
```

```python
df['duration_valor'] = df['duration'].fllna('0').str.split(' ').str[0].astype(int)
df
```

```python
panel = pyg.walk(df, dark='light')
```
